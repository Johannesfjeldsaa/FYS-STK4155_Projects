```python
from imageio.v2 import imread
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
```

```python
terrain = imread(r'DataFiles\SRTM_data_Norway_1.tif')

N = 500  # Number of datapoints to use in each direction

square = patches.Rectangle((10, 10), 480, 480, linewidth=2, edgecolor='red', facecolor='none')

# Create the left subplot for the original terrain with the square
plt.subplot(121)  # 1 row, 2 columns, select the first subplot
plt.imshow(terrain, cmap='gray')
plt.gca().add_patch(square)   
plt.title('Full terrain data')
plt.xlabel('X')
plt.ylabel('Y')

square = patches.Rectangle((1, 1), 497, 497, linewidth=2, edgecolor='red', facecolor='none')

# Create the right subplot for the terrain[:N, :N]
plt.subplot(122)  # 1 row, 2 columns, select the second subplot
plt.imshow(terrain[:N, :N], cmap='gray', vmin=terrain.min(), vmax=terrain.max())
plt.gca().add_patch(square)  
plt.title('Selected terrain data')
plt.xlabel('X')
plt.ylabel('Y')

plt.tight_layout()  # Adjust subplot layout for better appearance
plt.savefig("Benyttet_kartutsnitt.png")
plt.show()
```
